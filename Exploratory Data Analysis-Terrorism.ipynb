# Cleaning the Data

```python
Data=Data[["iyear","imonth","iday","country_txt","region_txt","provstate","city","latitude",
           "longitude","location","summary","attacktype1_txt","targtype1_txt","gname",
          "motive","weaptype1_txt","nkill","nwound","addnotes"]]
```

```python
Data
```

```python
Data.head()
```

```python
Data.rename(columns={"iyear":"Year","imonth":"Month","iday":"Day","country_txt":"Country",
                    "region_txt":"Region","provstate":"Province/State","city":"City",
                    "latitude":"Latitude","longitude":"Longitude","location":"Location",
                    "summary":"Summary","attacktype1_txt":"Attack Type","targtype1_txt":
                    "Target Type","gname":"Group Name","motive":"Motive","weaptype1_txt":
                    "Weapon Type","nkill":"Killed","nwound":"Wounded","addnotes":
                    "Add Notes"},inplace=True)
```

```python
Data.head()
```

```python
Data.info()
```

```python
Data.shape
```

```python
Data.isnull().sum()
```

```python
Data["Killed"]=Data["Killed"].fillna(0)
Data["Wounded"]=Data["Wounded"].fillna(0)
Data["Casualty"]=Data["Killed"]+Data["Wounded"]
```

```python
Data.describe()
```

# Visualize the Data

### 1. Year wise Attacks

#### 1.Number of Attacks in each Year

```python
attacks=Data["Year"].value_counts(dropna=False).sort_index().to_frame().rename(columns=
        {"index":"Year","Year":"Attacks"})
```

```python
attacks
```

```python
attacks.head()
```

```python
attacks.plot(kind="bar",color="blue",figsize=(15,6),fontsize=13).set(xlabel="Years",
             ylabel="Number of Attacks",title="Timeline of Attacks")
```

1.Most number of attacks occured in 2014=16903.
